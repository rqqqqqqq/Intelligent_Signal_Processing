Wave module will open the sound file. Mode rb means that it is in read only mode.

```python
audio = wave.open("Ex3_files/Ex3_sound1.wav", mode='rb')
```

### The following extracts the lsb and converts it into a string using the audio steganography method.

```python
# Byte array is being converted from the audio file
read_frames = audio.readframes(audio.getnframes())
byte_frame = bytearray(list(read_frames))

# LSB of each byte is extracted
lsb_extracted = [byte_frame[i] & 1 for i in range(len(byte_frame))]

# Byte array is being converted to string
message = "".join(chr(int("".join(map(str,lsb_extracted[i:i+8])),2)) for i in range(0,len(lsb_extracted),8))

# Cut off at filler characters
decoded_message = message.split("###")[0]

# Printing decoded text
print("This is the decoded message: " + decoded_message)
audio.close()
```
