## Extracting the secret message

Using urlib library to install thinkdsp module

```python
from thinkdsp import read_wave
```

Importing the necessary modules and libraries

```python
import numpy as np
import wave
from thinkdsp import read_wave
```

Playing the audio file before embedding message

```python
sound1 = read_wave("Ex3_files/Ex3_sound5.wav")
sound1.make_audio()
```

Wave module opens audio file. Mode = rb means that it is in read only mode.

```python
audio = wave.open("Ex3_files/Ex3_sound5.wav", mode='rb')
```

Read frames and convert to byte array

```python
read_frames = audio.readframes(audio.getnframes())
byte_frame = bytearray(list(read_frames))
```

Adding in the secret message into audio file

```python
# Preparing the secret text message
secret_message ='Father Christmas does not exist'

# Place dummy data into the rest of the bytes
new_message = secret_message + int((len(byte_frame)-(len(secret_message)*64))/8) *'#'

# Convert text to bit array
message_bits = list(map(int, ''.join([bin(ord(i)).lstrip('0b').rjust(8,'0') for i in new_message])))

# Placing one bit by one bit from the message into the LSB of each byte of audio data
for i, bit in enumerate(message_bits):
    byte_frame[i] = (byte_frame[i] & 254) | bit
    
# Retrieving the changed bytes
frame_modified = bytes(byte_frame)
```

Bytes are saved into a new audio file in wave format

```python
with wave.open('secret_message_audio.wav', 'wb') as fd:
    fd.setparams(audio.getparams())
    fd.writeframes(frame_modified)
audio.close()
```

Playing the new audio file with the embedded message

```python
sound2 = read_wave("secret_message_audio.wav")
sound2.make_audio()
```
